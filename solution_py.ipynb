```python
import sympy as sp

w, λ = sp.symbols('w λ')
x = sp.Matrix([1, sp.cos(w)]) / 2
k1, k2 = sp.exp(-w * sp.I) / 2, sp.exp(w * sp.I) / 2
mat = sp.Matrix([[1 / 2 + 0j, 1 / 2 + 0j], [k1, k2]])
P, D = mat.diagonalize()
P.simplify()
D.simplify()
```

```python
E = D ** 98
```

```python
mat = P * E * P.inv()
values = mat * x
values
```

```python
value = values[0] + values[1]
k = sp.symbols('k', integer=True)
dtft_under_integral = value * sp.exp(-1j * w * k) / 4 / sp.pi
open("under_integral.txt", "w").write(str(dtft_under_integral))
dtft_under_integral
```

```python
import sympy as sp
dtft_under_integral = sp.sympify(open("under_integral.txt").read())
integrated = sp.integrate(dtft_under_integral, (w, 0, sp.pi * 2))
integrated
```

```python
import numpy as np
N = 4
integral = 0
for i in range(N):
    integral += dtft_under_integral.evalf(subs={w: i * sp.pi / (N + 1), k: 0})
integral
```
