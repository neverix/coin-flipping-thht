```python
import sympy as sp

w, λ = sp.symbols('w λ')
x = sp.Matrix([1, sp.cos(w)])
k1, k2 = sp.exp(-w * sp.I) / 2, sp.exp(w * sp.I) / 2
mat = sp.Matrix([[1 / 2 + 0j, 1 / 2 + 0j], [k1, k2]])
P, D = mat.diagonalize()
P.simplify()
D.simplify()
```

```python
t = sp.symbols('t', integer=True)
E = D ** t
E.simplify()
```

```python
mat = P * E * P.inv()
mat.simplify()
mat
```

```python
values = mat * x
values.simplify()
values
```

```python
value = values[0] + values[1]
k = sp.symbols('k', integer=True)
dtft_under_integral = value * sp.exp(1j * w * k) / 2
dtft_under_integral.simplify()
open("under_integral.txt", "w").write(str(dtft_under_integral))
display(dtft_under_integral)
```
